```python
import matplotlib.pyplot as plt

df[['Postal Code','Country']]
df[['Country','City']]
df[['City','Order ID']]

df[20:]

df['Country'].value_counts()
df['Country'].unique()
df['Country'].nunique()

df['City'].unique()
df['City'].nunique()

df.loc[0]
df.loc[25,'Order Date']
df.loc[25,['Order Date','City']]

df.iloc[0:50,2]
df.iloc[0:50,[2,3,4]]
df.iloc[0:50,[14,17,20]]

df.drop_duplicates(inplace=True)

df['Order Date']=pd.to_datetime(df['Order Date'])

df['YEAR']=df['Order Date'].dt.year
df['MONTH']=df['Order Date'].dt.month
df['DATE']=df['Order Date'].dt.date
df['DAY']=df['Order Date'].dt.day

monthly_sales=df.groupby('MONTH')['Sales Profit'].sum()
display(monthly_sales)
monthly_sales.plot(kind='bar')
plt.title('Monthly Sales Profit')
plt.xlabel('Month')
plt.ylabel('Total Sales Profit')
plt.show()

monthly_sales.plot(kind='line')
plt.title('Monthly Sales Profit')
plt.xlabel('Month')
plt.ylabel('Total Sales Profit')
plt.show()

country_sales=df.groupby('State')['Sales Profit'].sum()
display(country_sales)
country_sales.plot(kind='line')
plt.title('Sales Profit by Country')
plt.xlabel('State')
plt.ylabel('Total Sales Profit')
plt.show()
country_sales.plot(kind='bar')
plt.title('Sales Profit by Country')
plt.xlabel('State')
plt.ylabel('Total Sales Profit')
plt.show()

product_sales=df.groupby('Clothing Type')['Sales Profit'].sum()
display(product_sales)

df.to_csv('cleaned_data.csv',index=False)

df_cleaned = df.copy()
```
